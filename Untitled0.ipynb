```python
%pylab inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from scipy.optimize import root
import matplotlib.pyplot as plt
import numpy as np
```

```python
iteraciones=1000000
omega=empty(0)
theta=empty(0)
k=empty(0)
gamma=1.5
m=2
n=2
alpha=3.2
beta=3.2
x=empty(iteraciones)
y=empty(iteraciones)
```

```python
k=append(k,0.2335433)
omega=append(omega,2.2)
theta=append(theta,2.1)
k,theta,omega
```

```python
for i in range(iteraciones):
  
    
    omeg = abs(np.random.normal(omega[i], 0.2))
    thet = abs(np.random.normal(theta[i], 0.2))
    k_0 = abs(np.random.normal(k[i], 0.2))
    def f(x):
        """Sistema de dos ecuaciones con dos incógnitas.
     
        """
        return [
            alpha+ (omeg/(1+((x[1]/k_0)**n)))-gamma*x[0],
            beta + (thet/(1+(x[0]/k_0)**m))-gamma*x[1]     
        ]
     
    sol = root(f, [1, 1], jac=False)  # Devuelve objecto Result
    f    
    x[i]=sol.x[0]
    y[i]=sol.x[1]
    a=(gamma**2)-((omeg*thet*n*m*(y[i]**(n-1))*(x[i]**(n-1)))/((((1+(y[i]/k_0)**n)*(1+(x[i]/k_0)**m))**2)*(k_0)**(m+n)))
    if(a>0) & (x[i]>0) & (y[i]>0):
        omega=append(omega,omeg)
        theta=append(theta,thet)
        k=append(k,k_0)
        
    if (a<0) & (x[i]>0) & (y[i]>0):
        omega=append(omega,omega[i])
        theta=append(theta,theta[i])
        k=append(k,k[i])
       
    if(x[i]<0)or (y[i]<0):
        omega=append(omega,omega[i])
        theta=append(theta,theta[i])
        k=append(k,k[i])
        
        
```
